```python
# Required imports for the entire module
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon
```

```python
def coin_toss_simulation(trials=10000):
    heads = 0
    tails = 0
    for _ in range(trials):
        outcome = random.choice(['H', 'T'])
        if outcome == 'H':
            heads += 1
        else:
            tails += 1
    print("Coin Toss Simulation:")
    print(f"Total Tosses: {trials}")
    print(f"Heads: {heads} ({heads / trials:.4f})")
    print(f"Tails: {tails} ({tails / trials:.4f})\n")
coin_toss_simulation()
```

```python
def dice_sum_seven_simulation(trials=10000):
    count_sum_7 = 0
    for _ in range(trials):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 + die2 == 7:
            count_sum_7 += 1
    print("Dice Roll Simulation:")
    print(f"Total Rolls: {trials}")
    print(f"Sum of 7 count: {count_sum_7}")
    print(f"Probability of sum 7: {count_sum_7 / trials:.4f}\n")

dice_sum_seven_simulation()
```

```python
def probability_at_least_one_six(trials=10000):
    success_count = 0
    for _ in range(trials):
        rolls = [random.randint(1,6) for _ in range(10)]
        if 6 in rolls:
            success_count += 1
    probability = success_count / trials
    print(f"Estimated P(at least one 6 in 10 rolls): {probability:.4f}\n")
probability_at_least_one_six()
```

```python
def conditional_probability_bayes_simulation(trials=1000):
    colors = ['R']*5 + ['G']*7 + ['B']*8 
    prev = None
    count_blue_before = 0
    blue_followed_by_red = 0

    for _ in range(trials):
        current = random.choice(colors)
        if prev == 'B':
            count_blue_before += 1
            if current == 'R':
                blue_followed_by_red += 1
        prev = current

    if count_blue_before > 0:
        cond_prob = blue_followed_by_red / count_blue_before
        print(f"P(Red | Blue before) = {cond_prob:.4f}")
    else:
        print("Not enough blue balls drawn for conditional probability.")
    p_r = 5 / 20
    p_b = 8 / 20
    p_b_given_r = 8 / 20
    bayes_prob = (p_b_given_r * p_r) / p_b
    print(f"Bayes Theorem result: P(Red|Blue) ≈ {bayes_prob:.4f}\n")

conditional_probability_bayes_simulation()
```

```python
def discrete_random_variable_simulation(sample_size=1000):
    values = [1, 2, 3]
    probabilities = [0.25, 0.35, 0.40]
    sample = np.random.choice(values, size=sample_size, p=probabilities)
    mean = np.mean(sample)
    variance = np.var(sample)
    std_dev = np.std(sample)
    print(f"Discrete RV Sample Statistics (n={sample_size}):")
    print(f"Mean: {mean:.4f}, Variance: {variance:.4f}, Std Dev: {std_dev:.4f}\n")
discrete_random_variable_simulation()
```

```python
def exponential_distribution_simulation(samples=2000, mean=5):
    scale = mean  # mean = scale for exponential
    data = np.random.exponential(scale=scale, size=samples)
    plt.figure(figsize=(10,6))
    plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogram')
    x = np.linspace(0, np.max(data), 1000)
    pdf = (1/scale) * np.exp(-x/scale)
    plt.plot(x, pdf, 'r-', lw=2, label='PDF (Exponential)')
    plt.title(f'Exponential Distribution (mean={mean}) - {samples} samples')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
exponential_distribution_simulation()
```

```python
def central_limit_theorem_simulation(total_samples=10000, sample_size=30, sample_count=1000):
    uniform_data = np.random.uniform(0, 1, total_samples)
    sample_means = [np.mean(np.random.choice(uniform_data, sample_size, replace=False)) for _ in range(sample_count)]
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.hist(uniform_data, bins=30, color='lightgreen', edgecolor='black')
    plt.title('Uniform Distribution (0,1)')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Sample Means Distribution (n={sample_size}, samples={sample_count})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
central_limit_theorem_simulation()
```
