```python
data=pd.read_csv('cleaned_austin_weather.csv')
data
```

```python
x=data.drop(['PrecipitationSumInches'],axis=1)
y=data['PrecipitationSumInches']
x
```

```python
y=y.values.reshape(-1,1)
y
```

```python
day_index=798
day=[i for i in range(y.size)]
day
```

```python
print('the precipitation trend graph')
plt.scatter(day,y,color='blue')
plt.scatter(day[day_index],y[day_index],color='red')
plt.title('precipittion level')
plt.xlabel('days')
plt.ylabel('precipitation in inches')
plt.show()
x_vis=x.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
import matplotlib.pyplot as plt
print('The precipitation vs attribute graphs')
for i in range(x_vis.shape[1]):
    plt.subplot(2, 3, i + 1)
    plt.scatter(day, x_vis.iloc[:, i], color='blue', s=10, label='All days')
    plt.scatter(day[day_index], x_vis.iloc[day_index, i], color='red', s=20, label='Selected days')
    plt.title(x_vis.columns[i])
    plt.xlabel("Day")
    plt.ylabel(x_vis.columns[i])
    plt.legend()
plt.tight_layout()
plt.show()
```
