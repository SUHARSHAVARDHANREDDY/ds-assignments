```python
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('cleaned_austin_weather.csv')
data
```

```python
x=data.drop(['PrecipitationSumInches'],axis=1)
y=data['PrecipitationSumInches']
```

```python
y=y.values.reshape(-1,1)
y
```

```python
day_index=798
day=[i for i in range(y.size)]
```

```python
clf=LinearRegression()
clf.fit(x,y)
```

```python
print('the precipitation trend graph')
plt.scatter(day,y,color='blue')
plt.scatter(day[day_index],y[day_index],color='red')
plt.title('precipittion level')
plt.xlabel('days')
plt.ylabel('precipitation in inches')
plt.show()
x_vis=x.filter(['TempavgF','DewpointavgF','HumidityavgPercent','SeaLevelPressureavgInches','VisibilityavgMiles','WindAvgMPH'])
```

```python
print('the precipitation vs attribute graph')
for i in range(x_vis.columns.size):
    plt.subplot(3,2,i+1)
    plt.scatter(day,x_vis[x_vis.columns.values[i][:100]],color='blue')
    plt.scatter(day[day_index],x_vis[x_vis.columns.values[i]][day_index],color='red')
    plt.title(x_vis.columns.values[i])

plt.show()
```
